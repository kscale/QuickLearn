# figure基本图像输出

```
import matplotlib.pyplot as plt
import numpy as np
```

```
x = np.linspace(-3, 3, 50) #使用np.linspace定义x：范围是(-3,3);个数是50.
y1 = 2*x + 1
y2 = x**2
```

```
plt.figure()
plt.plot(x, y1)
plt.show()
```

```
plt.figure(num=3,figsize=(8,5)) # 编号为3；大小为(8, 5)
plt.plot(x,y2)
plt.plot(x, y1, color='red', linewidth=1.0, linestyle='--')
plt.show()
```

# 设置坐标轴的名字与间隔

```
import matplotlib.pyplot as plt
import numpy as np
```

```
x=np.linspace(-3,3,50)
y1=2*x+1
y2=x**2
```

```
plt.figure()
plt.plot(x,y2)
plt.plot(x,y1,color='red',linewidth=1.0,linestyle='--')
```

```
plt.xlim((-1, 2))     #使用plt.xlim设置x坐标轴范围：(-1, 2)
plt.ylim((-2, 3))
plt.xlabel('I am x')
plt.ylabel('I am y')
plt.plot(x,y2)
plt.plot(x,y1)
plt.show()
```

```
new_ticks = np.linspace(-1, 2, 5)
print(new_ticks)
plt.xticks(new_ticks)
plt.yticks([-2, -1.8, -1, 1.22, 3],[r'$really\ bad$', r'$bad$', r'$normal$', r'$good$', r'$really\ good$'])
plt.ylim((-2, 3))  #可以对y轴范围进行调整
# 同样可以对x轴进行调整 plt.xlim((-1, 2)) 
plt.plot(x,y2)
plt.plot(x,y1)
plt.show()
```

# 调整坐标轴

```
x = np.linspace(-3, 3, 50)
y1 = 2*x + 1
y2 = x**2
```

```
plt.figure()
plt.plot(x, y2)
plt.plot(x, y1, color='red', linewidth=1.0, linestyle='--')
plt.xlim((-1, 2))
plt.ylim((-2, 3))

new_ticks = np.linspace(-1, 2, 5)
plt.xticks(new_ticks)
plt.yticks([-2, -1.8, -1, 1.22, 3],['$really\ bad$', '$bad$', '$normal$', '$good$', '$really\ good$'])
```

```
plt.figure()
plt.plot(x, y2)
plt.plot(x, y1, color='red', linewidth=1.0, linestyle='--')
plt.xlim((-1, 2))
plt.ylim((-2, 3))
ax = plt.gca()               #使用plt.gca获取当前坐标轴信息
ax.spines['right'].set_color('green')     #右边框
ax.spines['top'].set_color('none')      #上边框
ax.spines['bottom'].set_position(('data', 0)) #使用.set_position设置边框位置：y=0=0的位置
plt.show()
```

y轴移动到x=0的位置


```
plt.figure()
plt.plot(x, y2)
plt.plot(x, y1, color='red', linewidth=1.0, linestyle='--')
plt.xlim((-1, 2))
plt.ylim((-2, 3))
ax = plt.gca()               #使用plt.gca获取当前坐标轴信息
ax.spines['right'].set_color('green')     #右边框
ax.spines['top'].set_color('none')      #上边框
ax.spines['bottom'].set_position(('data', 0))#使用.set_position设置边框位置：y=0=0的位置
ax.spines['left'].set_position(('data', 0)) #y轴移动到x=0的位置
plt.show()
```

# Legend图例



plt.legend()使用

```
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-3, 3, 50)
y1 = 2*x + 1
y2 = x**2

plt.figure()
#set x limits
plt.xlim((-1, 2))
plt.ylim((-2, 3))

# set new sticks
new_sticks = np.linspace(-1, 2, 5)
plt.xticks(new_sticks)
# set tick labels
plt.yticks([-2, -1.8, -1, 1.22, 3],
           [r'$really\ bad$', r'$bad$', r'$normal$', r'$good$', r'$really\ good$'])


# set line syles
l1 = plt.plot(x, y1, label='linear line')
l2 = plt.plot(x, y2, color='red', linewidth=1.0, linestyle='--', label='square line')

#添加图例
plt.legend(loc='upper right')
```

# 画图种类

## Scatter散点图
生成1024个呈标准正态分布的二维数据组 (平均数是0，方差为1) 作为一个数据集，并图像化这个数据集。每一个点的颜色值用T来表示：

```
import matplotlib.pyplot as plt
import numpy as np

n = 1024    # data size
X = np.random.normal(0, 1, n) # 每一个点的X值
Y = np.random.normal(0, 1, n) # 每一个点的Y值
T = np.arctan2(Y,X) # for color value

plt.scatter(X, Y, s=75, c=T, alpha=.5)
#输入X和Y作为location，size=75，颜色为T，color map用默认值，透明度alpha 为 50%。
#x轴显示范围定位(-1.5，1.5)，并用xtick()函数来隐藏x坐标轴，y轴同理
plt.xlim(-1.5, 1.5)
plt.xticks(())  # ignore xticks
plt.ylim(-1.5, 1.5)
plt.yticks(())  # ignore yticks

plt.show()
```

## Bar柱状图

```
import matplotlib.pyplot as plt
import numpy as np

n = 12
X = np.arange(n)
Y1 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)
Y2 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)

plt.bar(X, +Y1)
plt.bar(X, -Y2)

plt.xlim(-.5, n)
plt.xticks(())
plt.ylim(-1.25, 1.25)
plt.yticks(())

plt.show()
```
